# 1
a) Faça um modelo de regressão linear simples utilizando a variável 'KmRodado' para prever a 'PrecoVenda'.<br>

b) Calcule o R2 para o modelo criado.

```python
from sklearn import linear_model

dfCarros = pd.read_csv('../data/datasetCarros.csv')

regr = linear_model.LinearRegression()

akm = np.array(dfCarros['KmRodado']).reshape(-1, 1)

regr.fit(akm, dfCarros['PrecoVenda'])

r_sq = regr.score(akm, dfCarros['PrecoVenda'])

print('R2: ', r_sq)
```

```python
print(dfCarros.info())
print(dfCarros['TipoCombustivel'].unique())
print(dfCarros['Trasmissao'].unique())
```

```python
dfCarros['TipoCombustivel'] = dfCarros['TipoCombustivel'].map({'Petrol': 0, 'Diesel': 1, 'CNG': 2}).astype('int8')
dfCarros['Trasmissao'] = dfCarros['Trasmissao'].map({'Manual': 0, 'Automatic': 1}).astype('int8')
```

```python
from sklearn.model_selection import train_test_split

dfCarros_train, dfCarros_test = train_test_split(dfCarros, test_size=0.5, random_state=10)

lm = linear_model.LinearRegression()

lm.fit(dfCarros_train.drop(columns=['Nome', 'PrecoVenda']).values, dfCarros_train['PrecoVenda'].values)
print(lm.intercept_, lm.coef_)
```

```python
from sklearn.linear_model import Ridge

regrRidge = Ridge(alpha=2.0)

regrRidge.fit(dfCarros_train.drop(columns=['Nome', 'PrecoVenda']).values, dfCarros_train['PrecoVenda'].values)

y_pred = regrRidge.predict(dfCarros_test.drop(columns=['Nome', 'PrecoVenda']).values)

pred = lm.predict(dfCarros_test.drop(columns=['Nome', 'PrecoVenda']).values)

print("Coeficientes:", regrRidge.coef_)
print("R2: %.4f" % regrRidge.score(dfCarros_test.drop(columns=['Nome', 'PrecoVenda']).values, dfCarros_test['PrecoVenda'].values))
```
